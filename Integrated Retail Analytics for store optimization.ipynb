# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Date features
merged_df['Month'] = merged_df['Date'].dt.month
merged_df['Week'] = merged_df['Date'].dt.isocalendar().week
merged_df['Year'] = merged_df['Date'].dt.year
# Encode categorical variables
merged_df['Type'] = merged_df['Type'].astype('category').cat.codes
```

```python
# Calculate IQR for Weekly_Sales
Q1 = merged_df['Weekly_Sales'].quantile(0.25)
Q3 = merged_df['Weekly_Sales'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for typical data
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_df = merged_df[(merged_df['Weekly_Sales'] >= lower_bound) & (merged_df['Weekly_Sales'] <= upper_bound)]

print("Original size:", merged_df.shape)
print("Filtered size:", filtered_df.shape)
```

```python
# Feature set
features = ['Temperature','Fuel_Price','Total_MarkDown','CPI','Unemployment',
            'IsHoliday','Type','Size','Month','Week','Year']
X = filtered_df[features]
y = filtered_df['Weekly_Sales']
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=38)
```

# 1st Model

```python
from sklearn.model_selection import GridSearchCV
# Define parameter grid
param_grid = {
    'n_estimators': [100, 200],           # number of trees
    'max_depth': [10, 20, None],          # tree depths
    'min_samples_split': [2, 5],          # minimum number of samples to split a node
    'min_samples_leaf': [1, 2],           # minimum number of samples in a leaf
    'max_features': ['sqrt', 'log2']      # number of features for best split
}

# Initialize model
rf = RandomForestRegressor(random_state=42)

# GridSearchCV setup
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)

# Fit to training data
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best hyperparameters:")
print(grid_search.best_params_)
print("Best RMSE on validation folds:", -grid_search.best_score_)

# Use the best model on your test data
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
test_rmse = mean_squared_error(y_test, y_pred) ** 0.5
print("Test set RMSE:", test_rmse)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # diagonal line
plt.xlabel('Actual Weekly Sales')
plt.ylabel('Predicted Weekly Sales')
plt.title('Predicted vs Actual Weekly Sales')
plt.grid(True)
plt.show()
```

```python
import lightgbm as lgb

train_data = lgb.Dataset(X_train, label=y_train)
valid_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.1,
    'verbose': -1,
    'random_state': 42
}

gbm = lgb.train(
    params,
    train_data,
    num_boost_round=200,
    valid_sets=[valid_data],
    callbacks=[lgb.early_stopping(stopping_rounds=20)]
)
y_pred_lgb = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

# Market Basket Analysis and Segmentation

```python
# We don't have any transaction and product data so market based analysis is not feasible with this dataset.
```

```python

from sklearn.cluster import KMeans
# Calculate Recency, Frequency, Monetary (RFM) values for each store
import datetime as dt

snapshot_date = merged_df['Date'].max() + pd.Timedelta(days=1)

rfm = merged_df.groupby('Store').agg({
    'Date': lambda x: (snapshot_date - x.max()).days,
    'Store': 'count',
    'Weekly_Sales': 'sum'
}).rename(columns={'Date': 'Recency', 'Store': 'Frequency', 'Weekly_Sales': 'Monetary'}).reset_index()

# Normalize RFM features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

# KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Visualize clusters
plt.figure(figsize=(10,6))
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Cluster', palette='Set1')
plt.title('Store Segments by Frequency and Monetary Value')
plt.show()

print(rfm.head())
```
